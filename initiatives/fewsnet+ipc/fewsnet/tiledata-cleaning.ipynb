```python
def join_columns(df, columns, new_name):
    df.rename(columns={columns[0]:new_name}, inplace=True)
    # for each column in columns except the first one join it to the new_name column
    for col in columns[1:]:
        # do a left join on df[new_name] and df[col]
        df[new_name] = df[new_name].combine_first(df[col])
        del df[col]
    return df
```
