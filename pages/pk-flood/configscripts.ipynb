```python
# if text and visual and notes column are nan delete row
df = df.dropna(subset=['Text', 'Visual', 'Notes'], how='all')
df
```

```python
# for each row print text
full_config = []
for index, row in df.iterrows():
     
    print( "{ id: 'ch"+row['Chapter']+"', alignment: 'left', hidden: false, title: '', image: '', description: '"+row['Text']+"', location: { center: [69.086, 26.067], zoom: 7, pitch: 55, bearing: -28 }, mapAnimation: 'flyTo', rotateAnimation: false, callback: '', onChapterEnter: [], onChapterExit: [] },")
```

```python
# save full_config to text file
with open('config.txt', 'w') as f:
    for item in full_config:
        f.write("%s " % item)
        
```
